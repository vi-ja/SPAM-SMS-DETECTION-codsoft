```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)

import joblib

plt.style.use("ggplot")
```

```python
#Load Dataset

df = pd.read_csv("spam.csv", encoding="latin-1")
```

```python
df.head()
```

```python
#Dataset Information]

print("Shape of Dataset :", df.shape)

df.info()
```

```python
#Remove Extra Columns

df = df[['v1','v2']]

df.columns = ['label','message']

df.head()
```

```python
#Check Missing Values

df.isnull().sum()
```

```python
#Remove Duplicate Messages

print("Duplicate Messages :", df.duplicated().sum())

df = df.drop_duplicates()

print("New Shape :", df.shape)
```

```python
#Check Class Distribution

df['label'].value_counts()
```

```python
#Graph 1 (Spam vs Ham Count)

plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x='label',
    hue='label',
    palette='Set2',
    legend=False
)

plt.title("Spam vs Ham Messages")
plt.xlabel("Category")
plt.ylabel("Number of Messages")

plt.show()
```

```python
#Graph 2 (Pie Chart)

plt.figure(figsize=(6,6))

df['label'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['skyblue','tomato']
)

plt.ylabel("")

plt.title("Spam vs Ham Distribution")

plt.show()
```

```python
#Calculate the length of each SMS

df['message_length'] = df['message'].apply(len)

df.head()
```

```python
#Graph 3 - Distribution of Message Length

plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x='message_length',
    bins=50,
    kde=True,
    color='royalblue'
)

plt.title("Distribution of SMS Message Length")
plt.xlabel("Message Length")
plt.ylabel("Count")

plt.show()
```

```python
#Graph 4 - Boxplot (Spam vs Ham)

plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='label',
    y='message_length',
    hue='label',
    palette='Set2',
    legend=False
)

plt.title("Message Length Comparison")
plt.xlabel("Message Type")
plt.ylabel("Length of Message")

plt.show()
```

```python
#Convert Labels into Numeric Values

df['label'] = df['label'].map({
    'ham':0,
    'spam':1
})

df.head()
```

```python
#Separate Features and Target

X = df['message']

y = df['label']

print(X.shape)
print(y.shape)
```

```python
#TF-IDF Vectorization

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=5000
)

X = vectorizer.fit_transform(X)

print("TF-IDF Shape :", X.shape)
```

```python
#Split Dataset into Train and Test

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Samples :", X_train.shape)
print("Testing Samples :", X_test.shape)
```

```python
#Train Naive Bayes Model

nb_model = MultinomialNB()

nb_model.fit(X_train, y_train)

print("Naive Bayes Model Trained Successfully")
```

```python
#Prediction & Accuracy

nb_prediction = nb_model.predict(X_test)

accuracy = accuracy_score(
    y_test,
    nb_prediction
)

print("Naive Bayes Accuracy :", round(accuracy*100,2),"%")
```

```python
#Classification Report & Confusion Matrix

print(classification_report(
    y_test,
    nb_prediction
))

cm = confusion_matrix(
    y_test,
    nb_prediction
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Naive Bayes Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

plt.show()
```

```python
#Train Logistic Regression Model

lr_model = LogisticRegression(max_iter=1000)

lr_model.fit(X_train, y_train)

print("Logistic Regression Model Trained Successfully")
```

```python
#Logistic Regression Prediction & Accuracy

lr_prediction = lr_model.predict(X_test)

lr_accuracy = accuracy_score(
    y_test,
    lr_prediction
)

print("Logistic Regression Accuracy :", round(lr_accuracy*100,2),"%")
```

```python
#Logistic Regression Report

print(classification_report(
    y_test,
    lr_prediction
))

cm_lr = confusion_matrix(
    y_test,
    lr_prediction
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm_lr,
    annot=True,
    fmt='d',
    cmap='Greens'
)

plt.title("Logistic Regression Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
#Train Support Vector Machine (SVM)

svm_model = LinearSVC()

svm_model.fit(X_train, y_train)

print("Support Vector Machine Model Trained Successfully")
```

```python
#SVM Prediction & Accuracy

svm_prediction = svm_model.predict(X_test)

svm_accuracy = accuracy_score(
    y_test,
    svm_prediction
)

print("SVM Accuracy :", round(svm_accuracy*100,2),"%")
```

```python
#SVM Classification Report

print(classification_report(
    y_test,
    svm_prediction
))

cm_svm = confusion_matrix(
    y_test,
    svm_prediction
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm_svm,
    annot=True,
    fmt='d',
    cmap='Oranges'
)

plt.title("Support Vector Machine Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
#Compare Model Accuracy

nb_accuracy = accuracy_score(
    y_test,
    nb_prediction
)

comparison = pd.DataFrame({

    "Model":[
        "Naive Bayes",
        "Logistic Regression",
        "Support Vector Machine"
    ],

    "Accuracy":[
        nb_accuracy,
        lr_accuracy,
        svm_accuracy
    ]

})

comparison
```

```python
#Graph 5 - Model Comparison

plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x="Model",
    y="Accuracy",
    hue="Model",
    palette="viridis",
    legend=False
)

plt.title("Accuracy Comparison of Machine Learning Models")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.ylim(0.90, 1.00)

for index, value in enumerate(comparison["Accuracy"]):
    plt.text(index, value + 0.002, f"{value:.3f}", ha="center")

plt.show()
```

```python
#Display Best Model

best_model = comparison.loc[
    comparison["Accuracy"].idxmax()
]

print(best_model)
```

```python
#Final Result

print("="*50)

print("Spam SMS Detection Project Completed Successfully")

print("="*50)

print("\nModel Comparison\n")

print(comparison)

print("\nBest Model :")

print(best_model["Model"])

print("\nAccuracy :")

print(round(best_model["Accuracy"]*100,2),"%")
```

```python
#Predict a Custom SMS

custom_sms = [
    "Congratulations! You have won a FREE iPhone. Click the link to claim your prize."
]

custom_vector = vectorizer.transform(custom_sms)

prediction = svm_model.predict(custom_vector)

if prediction[0] == 1:
    print("Prediction : SPAM")
else:
    print("Prediction : HAM")
```

```python
#Test Multiple SMS Messages

messages = [

    "Hey Vijay, are you coming to college today?",

    "WINNER!! Claim your cash prize now by clicking this link.",

    "Meeting has been postponed to tomorrow at 10 AM.",

    "Congratulations! You've won a lottery worth ₹10,00,000."

]

messages_vector = vectorizer.transform(messages)

predictions = svm_model.predict(messages_vector)

for msg, pred in zip(messages, predictions):

    print("-"*60)

    print("Message :", msg)

    if pred == 1:
        print("Prediction : SPAM")

    else:
        print("Prediction : HAM")
```

```python
#Save the Best Model

joblib.dump(svm_model, "spam_model.pkl")

print("spam_model.pkl saved successfully.")
```

```python
#Save TF-IDF Vectorizer

joblib.dump(vectorizer, "tfidf_vectorizer.pkl")

print("tfidf_vectorizer.pkl saved successfully.")
```

```python
#Predict Using Loaded Model

new_sms = [
    "Get a free vacation package now. Hurry up!"
]

new_sms_vector = loaded_vectorizer.transform(new_sms)

prediction = loaded_model.predict(new_sms_vector)

if prediction[0] == 1:

    print("Prediction : SPAM")

else:

    print("Prediction : HAM")
```

```python
#Graph 6 - Top 20 Important TF-IDF Words

feature_names = vectorizer.get_feature_names_out()

importance = np.asarray(X.sum(axis=0)).ravel()

top_indices = importance.argsort()[-20:]

plt.figure(figsize=(12,6))

plt.barh(
    feature_names[top_indices],
    importance[top_indices]
)

plt.title("Top 20 Important TF-IDF Words")

plt.xlabel("TF-IDF Score")

plt.show()
```
